```python
from datasets import load_dataset
```

```python
df_boolq_malay = load_dataset('wanadzhar913/boolq-malay', split='validation') \
                  .filter(lambda x: x["language"] == 'Malay') \
                  .take(200) \
                  .remove_columns(['split']) \
                  .to_pandas()

df_boolq_malay.head(3)
```

```python
df_boolq_malay.to_json('../datasets/for_presentation/boolq-malay-val-200.jsonl', lines=True, orient='records')
```

```python
df_boolq_english = load_dataset('wanadzhar913/boolq-malay', split='validation') \
                  .filter(lambda x: x["language"] == 'English') \
                  .take(200) \
                  .remove_columns(['split']) \
                  .to_pandas()

df_boolq_english.head(3)
```

```python
df_boolq_english.to_json('../datasets/for_presentation/boolq-eng-val-200.jsonl', lines=True, orient='records')
```

```python
df_fib_malay = load_dataset('wanadzhar913/fib-malay', split='train') \
                  .filter(lambda x: x["language"] == 'Malay') \
                  .take(200) \
                  .rename_column('summary', 'question') \
                  .to_pandas()

df_fib_malay.head(3)
```

```python
df_fib_malay.to_json('../datasets/for_presentation/fib-malay-val-200.jsonl', lines=True, orient='records')
```
